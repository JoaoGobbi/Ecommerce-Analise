```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
# Definir o diretório para salvar as figuras
FIGURES_PATH = Path('../reports/figures')
FIGURES_PATH.mkdir(parents=True, exist_ok=True)

# Carregar o dataset RFM com clusters
rfm = pd.read_csv('../data/processed/rfm_clusters.csv')

rfm.head()
```

Gráfico 1 — Distribuição dos Scores R, F, V

```python
plt.figure(figsize=(15,5))

for i, score in enumerate(['R_Score', 'F_Score', 'V_Score']):
    plt.subplot(1,3,i+1)
    sns.countplot(x=score, data=rfm, palette='Blues')
    plt.title(f'Distribuição do {score}')
    plt.xlabel(score)
    plt.ylabel('Contagem')

plt.tight_layout()
plt.savefig(FIGURES_PATH / 'rfm_scores_distribution.png')
plt.show()
```

Gráfico 2 — Clientes por Cluster

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Cluster', data=rfm, palette='Set2')
plt.title('Distribuição dos Clientes por Cluster')
plt.xlabel('Cluster')
plt.ylabel('Número de Clientes')

plt.savefig(FIGURES_PATH / 'clientes_por_cluster.png')
plt.show()
```

Gráfico 3 — Frequência vs Valor (colorido pelos clusters)

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Frequencia', y='Valor', hue='Cluster', data=rfm, palette='Set1')
plt.title('Clientes: Frequência vs Valor por Cluster')
plt.xlabel('Frequência')
plt.ylabel('Valor')

plt.savefig(FIGURES_PATH / 'scatter_frequencia_valor.png')
plt.show()
```
